```python
#13
# Convertir todas las variables categóricas restantes a numéricas en el DataFrame 'EmpleadosAttrition_reducido'
# Aplicando One-Hot Encoding a todas las columnas categóricas automáticamente sin especificar cada una

# Identificar las columnas categóricas restantes en 'EmpleadosAttrition_reducido'
categorical_columns = EmpleadosAttrition_reducido.select_dtypes(include=['object', 'category']).columns

# Aplicar One-Hot Encoding a todas las columnas categóricas restantes
EmpleadosAttrition_reducido = pd.get_dummies(EmpleadosAttrition_reducido, columns=categorical_columns, drop_first=True)

# Mostrar las primeras filas para verificar las conversiones
EmpleadosAttrition_reducido.head()
```

```python
#14 y 15
# Asegurarnos de que la columna 'Attrition' esté presente y en formato numérico en 'EmpleadosAttrition_reducido'
# Convertir 'Attrition' a formato numérico (1 para 'Yes', 0 para 'No') desde el DataFrame original
if 'Attrition' in EmpleadosAttrition_original.columns:
    EmpleadosAttrition_reducido['Attrition'] = EmpleadosAttrition_original['Attrition'].map({'Yes': 1, 'No': 0})

# Calcular las correlaciones de todas las variables respecto a 'Attrition'
correlaciones = EmpleadosAttrition_reducido.corr()['Attrition'].abs().sort_values(ascending=False)

# Seleccionar las variables con una correlación mayor o igual a 0.1
columnas_seleccionadas = correlaciones[correlaciones >= 0.1].index

# Crear el DataFrame 'EmpleadosAttritionFinal' con las variables seleccionadas
EmpleadosAttritionFinal = EmpleadosAttrition_reducido[columnas_seleccionadas].copy()

# Mostrar las primeras filas del nuevo DataFrame
EmpleadosAttritionFinal.head()
```

```python
#16
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Escalar las variables en EmpleadosAttritionFinal para asegurar que todas tengan una influencia equilibrada
scaler = StandardScaler()
EmpleadosAttritionFinal_scaled = scaler.fit_transform(EmpleadosAttritionFinal)

# Aplicar PCA al DataFrame escalado
pca = PCA()
EmpleadosAttritionPCA = pca.fit_transform(EmpleadosAttritionFinal_scaled)

# Mostrar los primeros valores de la primera columna
primer_componente = EmpleadosAttritionPCA[:, 0]
primer_componente[:5]
```

```python
#17
# Calcular la cantidad mínima de componentes principales que explican al menos el 80% de la varianza
varianza_acumulada = pca.explained_variance_ratio_.cumsum()
num_componentes = (varianza_acumulada >= 0.80).argmax() + 1

# Seleccionar el número mínimo de componentes que explican el 80% de la varianza
componentes_principales = EmpleadosAttritionPCA[:, :num_componentes]

# Crear un DataFrame con estos componentes y asignar nombres de columna como C0, C1, ...
columnas_componentes = {f'C{i}': componentes_principales[:, i] for i in range(num_componentes)}
EmpleadosAttritionFinal = EmpleadosAttritionFinal.assign(**columnas_componentes)

# Mostrar las primeras filas del DataFrame actualizado
EmpleadosAttritionFinal.head()
```
